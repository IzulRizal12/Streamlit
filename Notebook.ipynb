## Data Wrangling

### Gathering Data

```python
df_product = pd.read_csv('/content/products_dataset.csv')
df_order = pd.read_csv('/content/orders_dataset.csv')
df_orderitem = pd.read_csv('/content/order_items_dataset.csv')

merged_df = pd.merge(df_orderitem, df_order, on='order_id', how='inner')
df = pd.merge(merged_df, df_product, on='product_id', how='inner')
```

### Cleaning Data

```python
# Hapus kolom-kolom yang memiliki missing value yang besar, karena kolom tersebut tidak memengaruhi hasil analisis
kolom_hapus = ['product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'order_delivered_carrier_date', 'order_delivered_customer_date']
df.drop(kolom_hapus, axis=1, inplace=True)
```

```python
# Mengisi nilai yang hilang dengan median untuk kolom yang memiliki nilai numerik
kolom_numerik = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
for kolom in kolom_numerik:
    median = df[kolom].median()
    df[kolom].fillna(median, inplace=True)
```

```python
# Mengkonversi kolom 'order_approved_at' menjadi tipe datetime jika belum
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])

# Melakukan interpolasi linear untuk mengisi nilai yang hilang
df['order_approved_at'].interpolate(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

```python
kolom = ['price', 'freight_value', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
# Fungsi untuk membersihkan outlier menggunakan IQR
def clean_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_cleaned

# Membersihkan outlier untuk setiap kolom yang ditentukan
for col in kolom:
    df = clean_outliers(df, col)
```
